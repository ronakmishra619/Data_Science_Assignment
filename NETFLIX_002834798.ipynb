#Which month should a producer choose to release new content to avoid competition? (Identify the month with the least amount of new content added.)

##This heatmap visualizes the monthly content updates on Netflix from 2008 to 2021. The code for generating this heatmap involves several steps in data manipulation and visualization using Python libraries like Pandas and Matplotlib.

```python
netflix_date = netflix_shows[['date_added']].dropna()
netflix_date['year'] = netflix_date['date_added'].apply(lambda x : x.split(', ')[-1])
netflix_date['month'] = netflix_date['date_added'].apply(lambda x : x.lstrip().split(' ')[0])

month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'][::-1]
df = netflix_date.groupby('year')['month'].value_counts().unstack().fillna(0)[month_order].T
plt.figure(figsize=(10, 7), dpi=200)
plt.pcolor(df, cmap='afmhot_r', edgecolors='white', linewidths=2) # heatmap
plt.xticks(np.arange(0.5, len(df.columns), 1), df.columns, fontsize=7, fontfamily='serif')
plt.yticks(np.arange(0.5, len(df.index), 1), df.index, fontsize=7, fontfamily='serif')

plt.title('Netflix Contents Update', fontsize=12, fontfamily='calibri', fontweight='bold', position=(0.20, 1.0+0.02))
cbar = plt.colorbar()

cbar.ax.tick_params(labelsize=8)
cbar.ax.minorticks_on()
plt.show()
```

Data Preparation: The netflix_date DataFrame is created by selecting the 'date_added' column from the 'netflix_shows' DataFrame and dropping any missing values. It then creates two new columns: one for the year and another for the month, extracted from the 'date_added' column using lambda functions.
Grouping and Reshaping Data: It groups the data by 'year' and counts how many shows were added each month. This aggregated data is reshaped into a matrix (DataFrame) with months as rows and years as columns, using the value_counts and unstack functions. The missing values are filled with zeros, and the months are ordered from January to December.

Visualization: The heatmap is created using the pcolor function from Matplotlib. The colors represent the number of shows added each month for each year, with darker shades indicating higher numbers. The 'afmhot_r' colormap is used, giving the heatmap a fiery look.

Styling: The axis ticks are adjusted to correspond to the center of each cell, and fonts are stylized for readability. A colorbar is added as a legend to indicate the scale of the content updates.

Title and Display: The plot is titled 'Netflix Contents Update' with specific styling, and finally, it is displayed using plt.show().

This heatmap provides insight into the seasonality of content updates on Netflix. For instance, there appears to be a trend of increased updates around certain months which could coincide with peak viewing times or strategic content release schedules.

In exploratory data analysis, such visualizations are crucial for quickly identifying patterns, trends, and anomalies in the data which might not be apparent from raw numbers. They also form a solid foundation for further analysis, hypothesis testing, and predictive modeling in the realm of data science.

If the latest year 2019 is considered, January and December were the months when comparatively much less content was released.Therefore, these months may be a good choice for the success of a new release!

# Analysing IMDB ratings to get top rated movies on Netflix

```python
imdb_ratings=pd.read_csv('/content/IMDb ratings.csv',usecols=['weighted_average_vote'])
imdb_titles=pd.read_csv('/content/IMDb movies.csv', usecols=['title','year','genre'])
ratings = pd.DataFrame({'Title':imdb_titles.title,
                    'Release Year':imdb_titles.year,
                    'Rating': imdb_ratings.weighted_average_vote,
                    'Genre':imdb_titles.genre})
ratings.drop_duplicates(subset=['Title','Release Year','Rating'], inplace=True)
ratings.shape
```


Executing an inner join between the ratings dataset and the Netflix dataset to obtain content that is rated on IMDB and is also accessible on Netflix.

```python
ratings.dropna()
joint_data=ratings.merge(netflix_overall,left_on='Title',right_on='title',how='inner')
joint_data=joint_data.sort_values(by='Rating', ascending=False)
```


##The top 10 highest-rated movies currently on Netflix include:

```python
from collections import Counter

genres=list(netflix_movies['listed_in'])
gen=[]

for i in genres:
    i=list(i.split(','))
    for j in i:
        gen.append(j.replace(' ',""))
g=Counter(gen)
```


In this code snippet, we're importing the Counter class from the collections module. We're then creating a list called genres containing the values from the 'listed_in' column of the DataFrame netflix_movies.

We initialize an empty list called gen to store the individual genres. Then, we iterate through each entry in genres, splitting it by commas and converting it into a list of individual genres. For each genre in the sublist, we remove any spaces and append it to the gen list.

Finally, we use the Counter class to count the occurrences of each genre in the gen list, storing the results in the variable g. This allows us to determine the frequency of each genre in the Netflix dataset.

This process is useful for understanding the distribution of genres across the Netflix movie collection, providing insights into the variety of content available on the platform.

# Examining TV series data on Netflix.

```python
countries1={}
netflix_shows['country']=netflix_shows['country'].fillna('Unknown')
cou1=list(netflix_shows['country'])
for i in cou1:
    #print(i)
    i=list(i.split(','))
    if len(i)==1:
        if i in list(countries1.keys()):
            countries1[i]+=1
        else:
            countries1[i[0]]=1
    else:
        for j in i:
            if j in list(countries1.keys()):
                countries1[j]+=1
            else:
                countries1[j]=1
```

Here, we're initializing an empty dictionary called countries1 to store the counts of TV series by country.

We start by filling any missing values in the 'country' column of the DataFrame netflix_shows with 'Unknown'. Then, we create a list cou1 containing the values from the 'country' column.

We iterate through each entry in cou1, splitting it by commas to handle multiple countries listed for some TV series. If there's only one country listed, we directly increment its count in the countries1 dictionary. If there are multiple countries listed, we iterate through each country and update its count accordingly in the countries1 dictionary.

This process allows us to tally the occurrences of TV series by country, providing insights into the geographic distribution of content on Netflix.

```python
countries_fin1={}
for country,no in countries1.items():
    country=country.replace(' ','')
    if country in list(countries_fin1.keys()):
        countries_fin1[country]+=no
    else:
        countries_fin1[country]=no

countries_fin1={k: v for k, v in sorted(countries_fin1.items(), key=lambda item: item[1], reverse= True)}
```


In this code snippet, we're refining the counts of TV series by country.

We initialize an empty dictionary called countries_fin1 to store the aggregated counts of TV series by country.

We then iterate through each key-value pair in the countries1 dictionary, which contains the counts of TV series by country.

For each country, we remove any spaces in its name using replace(' ','') to ensure consistency.

If the country already exists in the countries_fin1 dictionary, we add the count of TV series to its existing count. If it doesn't exist, we create a new entry for the country with its count.

After iterating through all countries, we sort the countries_fin1 dictionary by the counts of TV series in descending order using a lambda function as the key argument in the sorted() function.

This sorting process enables us to identify the countries with the highest number of TV series available on Netflix.

# Content-Based Recommendation System


The TF-IDF (Term Frequency-Inverse Document Frequency) score represents the frequency of a word appearing in a document, adjusted based on how frequently it appears across all documents. This adjustment aims to diminish the influence of words that commonly occur in plot overviews, thereby reducing their impact on computing the ultimate similarity score.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#removing stopwords
tfidf = TfidfVectorizer(stop_words='english')

#Replace NaN with an empty string
netflix_overall['description'] = netflix_overall['description'].fillna('')

#Construct the required TF-IDF matrix by fitting and transforming the data
tfidf_matrix = tfidf.fit_transform(netflix_overall['description'])

#Output the shape of tfidf_matrix
tfidf_matrix.shape
```

There are about 16151 words described for the 6234 movies in this dataset.


In this context, the Cosine similarity score is employed due to its independence from magnitude and its relatively straightforward and efficient calculation process.

<img src='https://i.stack.imgur.com/36r1U.png'>

```python
# Import linear_kernel
from sklearn.metrics.pairwise import linear_kernel

# Compute the cosine similarity matrix
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
indices = pd.Series(netflix_overall.index, index=netflix_overall['title']).drop_duplicates()
```

```python
def get_recommendations(title, cosine_sim=cosine_sim):
    idx = indices[title]

    # Get the pairwsie similarity scores of all movies with that movie
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the movies based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most similar movies
    sim_scores = sim_scores[1:11]

    # Get the movie indices
    movie_indices = [i[0] for i in sim_scores]

    # Return the top 10 most similar movies
    return netflix_overall['title'].iloc[movie_indices]
```

In this code, we're utilizing the linear_kernel function from scikit-learn to compute the cosine similarity matrix. This matrix measures the pairwise similarity between all Netflix titles based on their TF-IDF (Term Frequency-Inverse Document Frequency) representations.

We then create a Series called indices, where the index is set to the title of each Netflix title and the values are their respective indices. This Series enables us to quickly retrieve the index of a title given its name.

The get_recommendations function takes a title as input and returns the top 10 most similar titles based on cosine similarity scores. It first retrieves the index of the input title from the indices Series, computes the similarity scores between the input title and all other titles, sorts the titles based on their similarity scores in descending order, selects the top 10 most similar titles (excluding the input title itself), retrieves their indices, and finally returns the titles corresponding to those indices.

This code forms the core of a content-based recommendation system for Netflix titles, allowing users to find similar titles based on the content of their chosen title.

This recommendation is just based on the Plot.
